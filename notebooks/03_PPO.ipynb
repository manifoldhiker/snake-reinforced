```python
import numpy as np
ACTION_CODES = [K_w, K_a, K_s, K_d]


class PLESnakeGymEnv(gym.Env):
    def __init__(self, config):
        self.game = init_ple_env()

        self.action_space = Discrete(4)
        self.observation_space = Box(low=0, high=255, shape=(64,64,3), dtype=np.uint8)

    def reset(self):
        self.game.reset_game()
        return self.game.getScreenRGB()

    def step(self, action):
        action_code = ACTION_CODES[action]
        reward = self.game.act(action_code)
        
        if reward > 0:
            reward *= 100
        reward /= 100
        
        observation = self.game.getScreenRGB()
        done = self.game.game_over()
        info = {'snake_len': get_snake_len(self.game)}
        
        return observation, reward, done, info

    def close(self):
        self.game.reset_game()
```

```python
env = PLESnakeGymEnv(None)
model = PPO("CnnPolicy", env, verbose=1)
```

```python
from stable_baselines3.common.env_checker import check_env
check_env(env)
```

```python
model.learn(total_timesteps=250_000)
```

```python
obs = env.reset()
while True:
    action, _states = model.predict(obs)
    obs, rewards, dones, info = env.step(action)
    # env.render()
    print(rewards.mean())
```
